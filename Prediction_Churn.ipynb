stacked bar for senior citizen

```python
churn_counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)
churn_percent = churn_counts.div(churn_counts.sum(axis=1), axis=0) * 100

ax = churn_percent.plot(kind='bar', stacked=True, figsize=(6, 4))

for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.text(x + width / 2, y + height / 2, f'{height:.1f}%', ha='center', va='center')

# Final formatting
plt.title("Churn by  SeniorCitizen(stacked bar chart)")
plt.ylabel("Percentage(%)")
plt.xlabel("SeniorCitizen")
plt.xticks(rotation=45)
plt.legend(title='Churn')
plt.tight_layout()

plt.show()
```

Log transformation is commonly used to reduce the skewness of data, especially when there is a large disparity between the values

```python
#perforn log transformations
df['MonthlyCharges'] = np.log(df['MonthlyCharges'])
df['TotalCharges'] = np.log(df['TotalCharges'])
```
